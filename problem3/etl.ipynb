```python
import json
```

```python
with open('/Users/bhavishya/Desktop/data.json', 'r') as f:
    data = json.load(f)

for program in data["programs"]:
    orchestra = program["orchestra"]
    season = program["season"]
    print(f"Orchestra: {orchestra}, Season: {season}")

    for concert in program["concerts"]:
        date = concert["Date"]
        venue = concert["Venue"]
        location = concert["Location"]
        event_type = concert["eventType"]
        print(f"  Date: {date}, Venue: {venue}, Location: {location}, Event Type: {event_type}")
```
